# TP02 Data visualisation

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
import seaborn as sns
titanic = sns.load_dataset("titanic")
iris = sns.load_dataset("iris")
```

# Univariate viewing

## quantitative

```python
# boxplot
sns.boxplot(x=titanic.age)
plt.show()
```

```python
# histgramme
sns.distplot(titanic.age.dropna(), bins=5, kde="True") 
plt.show()
```

## Qualitatif

```python
sns.countplot(titanic["class"])
plt.show()
```

```python
# Multivariate visualisation
```

```python
# Dispersion diagram
sns.scatterplot(
    x="sepal_length", y="sepal_width", data=iris
)
plt.show()
```

```python
sns.scatterplot(
    x="sepal_length", y="sepal_width", hue="species", data=iris
)
plt.show()
```

```python
sns.scatterplot(
    x="sepal_length", y="sepal_width", hue="species", size="petal_length", data=iris
)
plt.show()
```

```python
# Correlation diagram
iris_quant = iris.drop(columns="species")
corr = iris_quant.corr()
sns.heatmap(corr, square=True)
plt.show()
```

```python
# multiple mustache boxes
sns.boxplot(x="species", y="sepal_length", data=iris)
plt.show()
```

```python
# two dimensions qualitatives count
sns.countplot(x="class", hue="embark_town", data=titanic)
plt.show()
```

```python
g = sns.FacetGrid(titanic, col="class", row="sex") 
g.map(sns.distplot, "age")
plt.show()
```

# 3, Visualisation du jeu de donnée sy02-p2019.csv

```python
import pandas as pd
from IPython.display import display
```

```python
X = pd.read_csv("data/sy02-p2019.csv")
pd.options.display.max_rows = None
pd.options.display.max_columns = None
display(X)              
```

```python
X["Note médian"].dtype 
# The column corresponding to the median score is not recognized as quantitative because of the "ABS"
```

```python
X = pd.read_csv("data/sy02-p2019.csv", na_values="ABS")
X["Note médian"].dtype
```

```python
sns.countplot(x="Branche", hue="Semestre", data=X)
plt.show()
```

```python
X["Note ECTS"] = pd.Categorical(X["Note ECTS"], ordered=True)
sns.countplot(x="Branche", hue="Note ECTS", data=X)
plt.show()
```

```python
sns.distplot(X["Note médian"])
plt.show()
sns.distplot(X["Note final"])
plt.show()
sns.scatterplot(x="Note médian", y="Note final", data=X)
plt.show()
sns.jointplot(x="Note médian", y="Note final", data=X)
plt.show()
```

```python
# plot the mustache box of median and final exam
X1 = X.melt(
    value_vars=["Note médian", "Note final"], 
    var_name="Type", 
    value_name="Note"
)
sns.boxplot(x="Type", y="Note", data=X1)
plt.show()
```

```python
# bar figure of number of correction completed in median and final exam
X1 = X.rename(columns={"Correcteur médian": "médian", "Correcteur final": "final"})
X1 = X1.melt(
    id_vars=["Note médian", "Note final"], 
    value_vars=["médian", "final"], 
    var_name="Examen", 
    value_name="Correcteur"
)
X1["Note"] = np.where(X1["Examen"] == "médian", X1["Note médian"],X1["Note final"])
X1 = X1.drop(columns=["Note médian", "Note final"])
X1
```

```python
sns.countplot(x="Correcteur", hue="Examen", data=X1)
plt.show()
```
